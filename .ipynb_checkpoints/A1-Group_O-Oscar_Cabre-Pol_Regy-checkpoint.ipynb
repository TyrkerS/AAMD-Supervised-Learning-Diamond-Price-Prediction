```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from sklearn.preprocessing import StandardScaler
from scipy.stats import zscore
from IPython.display import display
```

## 1. Lectura de les dades

Carreguem els fitxers `diamonds-train.csv` i `diamonds-test.csv`

```python
train = pd.read_csv('data/diamonds-train.csv')
test = pd.read_csv('data/diamonds-test.csv')

print("TRAIN:")
print("Dimensions train:", train.shape)
display(train.head())

print("TEST:")
display(test.head())
```

---
## 2. Pre-processament de les dades

### a. Identificar els patrons amb dades “no físiques”

En aquest pas detectem i eliminem els diamants amb valors físicament impossibles o incoherents. Aquests valors poden aparèixer per errors de mesura o de registre de dades, i poden perjudicar el rendiment dels models d'aprenentatge. Els criteris que utilitzem són els següents:

- Un diamant no pot tenir valors iguals o menors que zero als atributs `carat`, `x`, `y`, `z`.
- L'atribut `depth` (profunditat relativa) ha de trobar-se dins del rang [43, 79].
- L'atribut `table` (amplada del cim relatiu) ha de trobar-se dins del rang [43, 95].
- El valor mínim vàlid del preu (`price`) és $326. Valors inferiors s'han de considerar erronis.

Les files que no compleixen aquestes condicions s’eliminen del conjunt d’entrenament, ja que no són útils per a l’entrenament dels models.

```python
# Eliminem files amb valors físicament incorrectes
train = train[
    (train['carat'] > 0) &
    (train['x'] > 0) & (train['y'] > 0) & (train['z'] > 0) &
    (train['depth'] >= 43) & (train['depth'] <= 79) &
    (train['table'] >= 43) & (train['table'] <= 95) &
    (train['price'] >= 326)
]

test = test[
    (test['carat'] > 0) &
    (test['x'] > 0) & (test['y'] > 0) & (test['z'] > 0) &
    (test['depth'] >= 43) & (test['depth'] <= 79) &
    (test['table'] >= 43) & (test['table'] <= 95) &
    (test['price'] >= 326)
]

# Mostrem les dimensions després de netejar
print("Dimensions després de netejar train:", train.shape)
print("Dimensions després de netejar test:", test.shape)
```

### b. Identificar outliers a partir de la distribució de cada atribut. 

En aquest pas analitzem la distribució dels atributs numèrics per detectar valors que poden ser considerats outliers, és a dir, punts molt allunyats de la majoria de dades.

Per fer-ho, utilitzem el mètode basat en l’interval interquartílic (IQR), que defineix un valor com a outlier si es troba fora del rang següent:

- `[Q1 - 1.5 · IQR, Q3 + 1.5 · IQR]`

On:
- **Q1** és el primer quartil (25% més petit),
- **Q3** és el tercer quartil (75% més petit),
- **IQR** és l’interval interquartílic: `IQR = Q3 - Q1`.

S'aplica aquest criteri als atributs: `carat`, `x`, `y`, `z`, `depth`, `table` i `price`.  
A més, es generen gràfics de caixa (*boxplots*) per visualitzar clarament on es troben aquests valors extrems dins de cada distribució.


```python
def detectar_outliers_iqr(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    return df[(df[col] < Q1 - 1.5 * IQR) | (df[col] > Q3 + 1.5 * IQR)]

# Llista d'atributs numèrics a analitzar
atributs = ['carat', 'depth', 'table', 'x', 'y', 'z', 'price']

# Recollim outliers per cada atribut
outliers_per_atribut = {}

for col in atributs:
    outliers = detectar_outliers_iqr(train, col)
    outliers_per_atribut[col] = outliers
    print(f"{col}: {len(outliers)} outliers")

    # Dibuixem el boxplot per visualitzar els outliers
    plt.figure(figsize=(6, 1.5))
    sns.boxplot(x=train[col], color='skyblue')
    plt.title(f'Boxplot de {col}')
    plt.xlabel(col)
    plt.show()

# Unim tots els outliers i eliminem duplicats
outliers_totals = pd.concat(outliers_per_atribut.values()).drop_duplicates()

# Resultat final
print(f"\nNombre total d’outliers únics detectats per IQR: {len(outliers_totals)}")
```

### c. Identificar outliers a partir de scatter plots atribut/preu

En aquest pas visualitzem la relació entre cada atribut numèric i el preu (`price`) mitjançant scatter plots. Això ens permet detectar visualment possibles outliers que no segueixen la tendència general de les dades.

A diferència del mètode anterior basat en la distribució d’un sol atribut (IQR), aquí observem la coherència entre variables. Per exemple, un diamant amb molt pes (`carat`) i un preu molt baix pot ser sospitós, igual que un diamant petit amb un preu desproporcionadament alt.

Aquest tipus de gràfics ens ajuda a detectar valors que no serien tècnicament outliers segons la seva distribució individual, però que **trenquen la correlació esperada amb el preu**.


```python
# Atributs que analitzarem respecte al preu
atributs = ['carat', 'depth', 'table', 'x', 'y', 'z']

# Diccionari per guardar outliers visuals
outliers_visuals = {}

# Threshold de desviació per considerar un punt com a outlier
z_score_threshold = 3

# Scatter plots + detecció de punts "estranys" respecte al patró atribut vs preu
for attr in atributs:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(data=train, x=attr, y='price', alpha=0.3)
    plt.title(f'Scatter plot de {attr} vs price')
    plt.xlabel(attr)
    plt.ylabel('price')
    plt.grid(True)
    plt.show()

    # Calculem el z-score del preu dins de cada grup d’atribut per trobar valors fora de tendència
    z_attr = zscore(train[attr])
    z_price = zscore(train['price'])

    # Detectem punts que tenen alta desviació tant en l’atribut com en el preu
    mask = (abs(z_attr) > z_score_threshold) | (abs(z_price) > z_score_threshold)
    outliers_visuals[attr] = train[mask]

    print(f"{attr}: {len(outliers_visuals[attr])} outliers visuals detectats")
```

### d. Fer una taula amb els identificadors dels outliers, els seus atributs, i les raons per la qual cadascú es considera outlier.

Un cop detectats els outliers mitjançant dos mètodes (distribució per IQR i scatter plots atribut/preu), construïm una taula que reculli:

- L’**identificador** (`id`) de cada diamant detectat com a outlier.
- Els **valors dels atributs numèrics** corresponents.
- Les **raons** per les quals es considera un outlier, indicant si ha estat detectat per:
  - IQR (distribució individual)
  - Scatter (comportament anòmal respecte al preu)

Aquesta informació ens permetrà decidir posteriorment què fer amb aquests valors (eliminar-los, corregir-los o conservar-los).


```python
# Inicialitzem columna 'raons' buida
train['raons'] = ''

# Umbral per detectar outliers per z-score (scatter)
z_score_threshold = 3

# --- 1. Outliers per IQR amb detall ---
atributs = ['carat', 'depth', 'table', 'x', 'y', 'z', 'price']
for col in atributs:
    Q1 = train[col].quantile(0.25)
    Q3 = train[col].quantile(0.75)
    IQR = Q3 - Q1
    low = Q1 - 1.5 * IQR
    high = Q3 + 1.5 * IQR

    mask_low = train[col] < low
    mask_high = train[col] > high

    train.loc[mask_low, 'raons'] += f'IQR:{col}<min; '
    train.loc[mask_high, 'raons'] += f'IQR:{col}>max; '

# --- 2. Outliers per Scatter plot (z-score) amb detall ---
z_price = zscore(train['price'])
atributs = ['carat', 'depth', 'table', 'x', 'y', 'z']
for col in atributs:
    z_attr = zscore(train[col])
    mask_attr = abs(z_attr) > z_score_threshold
    mask_price = abs(z_price) > z_score_threshold
    mask_combined = mask_attr | mask_price

    train.loc[mask_combined, 'raons'] += f'Scatter:{col}; '

# --- 3. Crear taula d'outliers amb raons i mostrar ---
taula_outliers = train[train['raons'] != ''].copy()
taula_outliers = taula_outliers[['id', 'carat', 'depth', 'table', 'x', 'y', 'z', 'price', 'raons']]
taula_outliers = taula_outliers.sort_values(by='id').reset_index(drop=True)

# Mostrem les primeres files de la taula

display(taula_outliers.head(10))
display(taula_outliers.tail(10))
print("Dimensions:",  taula_outliers.shape) 
```

### e. Decidir què fer amb els outliers i aplicar-ho als conjunts de dades

Un cop identificats els outliers amb detall, cal decidir si s’han d’eliminar o conservar. En aquest cas, optem per **eliminar-los** per evitar que afectin negativament l’ajust dels models de predicció i classificació.

Aquesta neteja s’aplica només sobre el conjunt d'entrenament (`train`), ja que el conjunt de test (`test`) ha de quedar intacte per poder fer una avaluació objectiva dels models.

S'eliminaran tots els registres que hagin estat marcats amb alguna raó a la columna `raons`.


```python
# Guardem el nombre original de files
n_original = len(train)

# Eliminem els outliers identificats (per id)
ids_outliers = taula_outliers['id']
train = train[~train['id'].isin(ids_outliers)].copy()

# Eliminem la columna 'raons' perquè ja no cal
if 'raons' in train.columns:
    train.drop(columns='raons', inplace=True)

# Resultat de la neteja
print(f"Número de files abans de netejar: {n_original}")
print(f"Número de files eliminades: {len(ids_outliers)}")
print(f"Número de files després de netejar: {len(train)}")

display(train.head(10))
display(train.tail(10))
print("Dimensions:",  train.shape) 
```

### f. Convertir els atributs categòrics en numèrics

Els atributs `cut`, `color` i `clarity` són variables categòriques amb **ordre semàntic**, és a dir, tenen valors que poden ordenar-se de pitjor a millor qualitat. Per tant, **no s’ha d’utilitzar one-hot encoding**, sinó una codificació numèrica **ordinal**, que conservi aquest ordre.

Codifiquem els valors d’aquesta manera:

- **cut:** Fair < Good < Very Good < Premium < Ideal
- **color:** J < I < H < G < F < E < D
- **clarity:** I1 < SI2 < SI1 < VS2 < VS1 < VVS2 < VVS1 < IF

```python
# Diccionaris d'ordenació
ordre_cut = {
    'Fair': 0,
    'Good': 1,
    'Very Good': 2,
    'Premium': 3,
    'Ideal': 4
}

ordre_color = {
    'J': 0,
    'I': 1,
    'H': 2,
    'G': 3,
    'F': 4,
    'E': 5,
    'D': 6
}

ordre_clarity = {
    'I1': 0,
    'SI2': 1,
    'SI1': 2,
    'VS2': 3,
    'VS1': 4,
    'VVS2': 5,
    'VVS1': 6,
    'IF': 7
}

# Aplicació al conjunt d'entrenament i test
train['cut'] = train['cut'].map(ordre_cut)
train['color'] = train['color'].map(ordre_color)
train['clarity'] = train['clarity'].map(ordre_clarity)

test['cut'] = test['cut'].map(ordre_cut)
test['color'] = test['color'].map(ordre_color)
test['clarity'] = test['clarity'].map(ordre_clarity)

# Comprovació de la conversió
print(train[['cut', 'color', 'clarity']].head())
print(test[['cut', 'color', 'clarity']].head())
```

### g. Estandaritzar tots els atributs per separat, excepte x, y, z i price

L'estandardització transforma les variables perquè tinguin mitjana 0 i desviació estàndard 1. Això és especialment important per a mètodes com la regressió o k-NN, que són sensibles a l’escala dels valors.

En aquest pas **només s’estandarditzen els atributs numèrics** que no són:
- `x`, `y`, `z` (que s’escaparan conjuntament en el pas 2.h)
- `price` (no es toca perquè després es transformarà amb logaritme)

L’estandardització es fa utilitzant la mitjana i desviació del conjunt `train`, i després s’aplica també a `test` per coherència.


```python
# Atributs que s’han d’estandarditzar (excloem 'x', 'y', 'z', 'price', 'id')
atributs_a_estandaritzar = ['carat', 'depth', 'table', 'cut', 'color', 'clarity']

# Inicialitzem l'estandarditzador
scaler = StandardScaler()

# Estandardització del conjunt d'entrenament
train[atributs_a_estandaritzar] = scaler.fit_transform(train[atributs_a_estandaritzar])

# Apliquem la mateixa transformació al conjunt de test
test[atributs_a_estandaritzar] = scaler.transform(test[atributs_a_estandaritzar])

# Comprovació
train[atributs_a_estandaritzar].describe().round(2)

print("TRAIN:")
print("Dimensions train:", train.shape)
display(train.head())

print("TEST:")
print("Dimensions test:", test.shape)
display(test.head())
```

### h. Escalar x, y i z entre -1 i 1, però de forma conjunta

En aquest pas normalitzem els valors de `x`, `y` i `z` perquè estiguin dins de l’interval [-1, 1], però **considerant els tres atributs conjuntament**, no per separat.

Això vol dir que busquem el valor mínim i màxim **global** entre les tres columnes, i després apliquem una transformació lineal comuna a tots tres atributs.

Aquesta normalització fa que les dimensions físiques tinguin la mateixa escala i no afectin de forma desigual els models.


```python
# 1. Juntem els valors de x, y i z per trobar mínim i màxim global
xyz_train = pd.concat([train['x'], train['y'], train['z']])
xyz_test = pd.concat([test['x'], test['y'], test['z']])
xyz_total = pd.concat([xyz_train, xyz_test])

min_xyz = xyz_total.min()
max_xyz = xyz_total.max()

# 2. Funció per escalar entre -1 i 1
def escalar_conjunt_entre_m1_i_1(valor):
    return 2 * (valor - min_xyz) / (max_xyz - min_xyz) - 1

# 3. Apliquem la funció als tres atributs, tant en train com en test
for col in ['x', 'y', 'z']:
    train[col] = escalar_conjunt_entre_m1_i_1(train[col])
    test[col] = escalar_conjunt_entre_m1_i_1(test[col])

# 4. Comprovació del rang
print("Rang x:", train['x'].min(), "→", train['x'].max())
print("Rang y:", train['y'].min(), "→", train['y'].max())
print("Rang z:", train['z'].min(), "→", train['z'].max())

print("TRAIN:")
print("Dimensions train:", train.shape)
display(train.head())

print("TEST:")
print("Dimensions test:", test.shape)
display(test.head())
```

### i. Transformació logarítmica del preu i escalat entre 0.1 i 0.9

El preu (`price`) dels diamants té un rang molt ampli (de $326 fins a gairebé $19,000), fet que pot perjudicar els models de regressió.

Per això, en lloc de treballar amb el preu directament, es calcula el **logaritme natural del preu**, que redueix la variabilitat i fa la distribució més propera a la normal.

Després, aquest logaritme es **rescalarà linealment** entre **0.1 i 0.9** per normalitzar-lo. Aquesta nova variable s’anomenarà `log_price`.

Aquesta transformació només es fa servir durant l’entrenament; per calcular l’error final (MAPE), es tornaran a usar els preus originals.


```python
# 1. Calculem el logaritme natural del preu
train['log_price'] = np.log(train['price'])
test['log_price'] = np.log(test['price'])

# 2. Calculem mínim i màxim del log_price sobre train
min_log = train['log_price'].min()
max_log = train['log_price'].max()

# 3. Funció per escalar entre 0.1 i 0.9
def escalar_log(x):
    return 0.1 + 0.8 * (x - min_log) / (max_log - min_log)

# 4. Apliquem l'escalat
train['log_price'] = escalar_log(train['log_price'])
test['log_price'] = escalar_log(test['log_price'])

# 5. Comprovació del resultat
print(train['log_price'].describe())

print("TRAIN:")
print("Dimensions train:", train.shape)
display(train.head())

print("TEST:")
print("Dimensions test:", test.shape)
display(test.head())
```

### j. Construcció de l’atribut binari `high_price`

Per a la part de classificació de la pràctica, cal una variable objectiu binària que indiqui si un diamant té **preu alt** o no.

Es defineix l’atribut `high_price` com:

- `1` si el `price` és **igual o superior a $2,500**
- `0` en cas contrari

Aquesta nova variable s’utilitzarà com a etiqueta per als models de classificació.


```python
# 1. Creem l’atribut binari segons el valor del preu
train['high_price'] = (train['price'] >= 2500).astype(int)
test['high_price'] = (test['price'] >= 2500).astype(int)

# 2. Comprovació de la distribució de classes
print(train['high_price'].value_counts())
print(test['high_price'].value_counts())

print("TRAIN:")
print("Dimensions train:", train.shape)
display(train.head())

print("TEST:")
print("Dimensions test:", test.shape)
display(test.head())
```

### Conclusió de la fase de pre-processament

En aquesta fase hem transformat el conjunt de dades inicial en un format òptim per a l’entrenament de models supervisats, aplicant diversos passos sistemàtics:

- S’han eliminat valors físicament impossibles (com dimensions iguals a 0).
- S’han detectat i eliminat outliers tant per distribució (IQR) com per relació amb el preu (scatter plots).
- Els atributs categòrics ordenats (`cut`, `color`, `clarity`) s’han codificat numèricament de forma ordinal.
- Els atributs numèrics han estat estandarditzats, i les dimensions físiques (`x`, `y`, `z`) s’han normalitzat conjuntament entre -1 i 1.
- El preu s’ha transformat amb logaritme i escalat entre 0.1 i 0.9, i s’ha creat la variable binària `high_price` per a classificació.

Aquest pre-processament assegura que les dades tinguin **qualitat, coherència i escala adequada** per obtenir models predictius i classificadors robustos i precisos.


```python
# Exportem els conjunts pre-processats a CSV
train.to_csv("train-preprocessed.csv", index=False)
test.to_csv("dtest-preprocessed.csv", index=False)
```
